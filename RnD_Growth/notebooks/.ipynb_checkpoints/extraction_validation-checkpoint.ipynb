```python
import requests
import pandas as pd
from pathlib import Path

INDICATORS = {
    "rd_gdp": "GB.XPD.RSDV.GD.ZS",
    "gdp_growth": "NY.GDP.MKTP.KD.ZG",
    "population": "SP.POP.TOTL",
    "tertiary_enrollment": "SE.TER.ENRR"
}

BASE_URL = "https://api.worldbank.org/v2/country/all/indicator"

records = []

for name, indicator in INDICATORS.items():

    page = 1
    
    while True:
        url = f"{BASE_URL}/{indicator}"
        
        params = {
            "date": "2005:2024",
            "format": "json",
            "per_page": 20000,
            "page": page
        }

        r = requests.get(url, params=params)
        data = r.json()

        if len(data) < 2:
            break

        rows = data[1]

        if not rows:
            break

        for row in rows:
            records.append({
                "country": row["country"]["id"],
                "year": row["date"],
                "indicator": name,
                "value": row["value"]
            })

        page += 1


df = pd.DataFrame(records)

print(df.shape)
df.head()
```

#### This code gave proper results

```python
import requests
import pandas as pd

INDICATORS = {
    "rd_gdp": "GB.XPD.RSDV.GD.ZS",
    "gdp_growth": "NY.GDP.MKTP.KD.ZG",
    #"population": "SP.POP.TOTL",
    #"tertiary_enrollment": "SE.TER.ENRR"
}

BASE_URL = "https://api.worldbank.org/v2/country/US/indicator"

records = []

for name, code in INDICATORS.items():
    url = f"{BASE_URL}/{code}"
    
    params = {
        "date": "2005:2024",
        "format": "json",
        "per_page": 100, 
        "page": 1
    }

    try:
        response = requests.get(url, params=params)
        data = response.json()
        if len(data) > 1:
            for item in data[1]:
                if item["value"] is not None:  
                    records.append({
                        "country": item["country"]["value"],
                        "year": int(item["date"]),
                        "indicator": name,
                        "value": item["value"]
                    })
    except Exception as e:
        print(f"Error fetching {name}: {e}")

df = pd.DataFrame(records)

if not df.empty:
    df_pivot = df.pivot(index=['country', 'year'], columns='indicator', values='value').reset_index()
    print(df_pivot.tail())
else:
    print("No data found.")
```

```python
import requests
import pandas as pd

INDICATORS = {
    "rd_gdp": "GB.XPD.RSDV.GD.ZS",
    "gdp_growth": "NY.GDP.MKTP.KD.ZG",
    #"population": "SP.POP.TOTL",
    #"tertiary_enrollment": "SE.TER.ENRR"
}

BASE_URL = "https://api.worldbank.org/v2/country/DE;JP;KR/indicator"

records = []

for name, code in INDICATORS.items():
    url = f"{BASE_URL}/{code}"
    
    params = {
        "date": "2005:2022",
        "format": "json",
        "per_page": 100, 
        "page": 1
    }

    try:
        response = requests.get(url, params=params)
        data = response.json()
        if len(data) > 1:
            for item in data[1]:
                if item["value"] is not None:  
                    records.append({
                        "country": item["country"]["value"],
                        "year": int(item["date"]),
                        "indicator": name,
                        "value": item["value"]
                    })
    except Exception as e:
        print(f"Error fetching {name}: {e}")

df = pd.DataFrame(records)

if not df.empty:
    df_pivot = df.pivot(index=['country', 'year'], columns='indicator', values='value').reset_index()
    print(df_pivot.tail())
else:
    print("No data found.")
```

```python
import requests
import pandas as pd
from io import StringIO

URL = "https://sdmx.oecd.org/public/rest/data/OECD.STI.STP,DSD_MSTI@DF_MSTI,/.A.B+GV+G+T_RS...?startPeriod=1981&endPeriod=2024&dimensionAtObservation=AllDimensions"

response = requests.get(url)
df = pd.read_csv(StringIO(response.text))
print(df.shape)
print(df.head(20))
```

```python
print(df['MEASURE'].unique())
```

```python
MY_MEASURES = ['G','B','GV','T_RS']
df_filtered = df[df['MEASURE'].isin(MY_MEASURES)]
print(df_filtered.shape)
print(df_filtered[['REF_AREA', 'MEASURE', 'TIME_PERIOD', 'OBS_VALUE']].head(20))
```

```python
df_clean = df_filtered[['REF_AREA', 'MEASURE', 'TIME_PERIOD', 'OBS_VALUE']].copy()
print(df_clean.shape)
print(df_clean.head(20))
```

```python
import requests, os

FILES = {
    "g_patent.tsv.zip": "https://s3.amazonaws.com/data.patentsview.org/download/g_patent.tsv.zip",
    "g_inventor_disambiguated.tsv.zip": "https://s3.amazonaws.com/data.patentsview.org/download/g_inventor_disambiguated.tsv.zip",
}

os.makedirs("../raw_data", exist_ok=True)

for fname, url in FILES.items():
    dest = f"../raw_data/{fname}"
    if os.path.exists(dest):
        print(f"Already exists: {dest}")
        continue
    print(f"Downloading {fname}...")
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(dest, "wb") as f:
            for chunk in r.iter_content(chunk_size=8 * 1024 * 1024):
                f.write(chunk)
    print(f"  Saved  ")
```
