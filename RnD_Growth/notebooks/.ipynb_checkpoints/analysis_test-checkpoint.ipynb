### Does national RnD spending correlate to patent output? 

#### Loading data to NOTEBOOK:

```python
import pandas as pd
df_B = pd.read_parquet('../processed/analytical_data/subset_B.parquet')
print('shape_B =', df_B.shape)
df_C = pd.read_parquet('../processed/analytical_data/subset_C.parquet')
print('shape_C =', df_C.shape)
```

#### Data quality checking

```python
print('B =', df_B.columns)
print('null =', df_B.isna().sum())
print('\nC =', df_C.columns)
print('null =', df_C.isna().sum())
```

- We'll use subset_B for this analysis as it has better rd_gdp data.

```python
df_B_agg = df_B.groupby(['country_code', 'year'], observed=True).agg({
    'patent_count': 'sum',          
    'inventor_count': 'sum',         
    'rd_gdp': 'first',               
    'gdp_growth': 'first',           
    'population': 'first',
    'tertiary_enrollment': 'first'
}).reset_index()
print(df_B_agg.columns)
df_B_agg.shape
```

#### Statistical Analysis application to the data for checking correlations and finding patterns. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_B_agg, x='rd_gdp', y='patent_count', 
                alpha=0.7, edgecolor='w', s=60)
plt.xlabel('R&D Spending (% of GDP)')
plt.ylabel('Total Patent Count')
plt.title('R&D Spending vs Patent Output (Raw)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

r, p_value = pearsonr(df_B_agg['rd_gdp'], df_B_agg['patent_count'])

print(f"   Pearson r = {r:.3f}")
print(f"   p-value   = {p_value:.3f}")
if p_value < 0.05:
    print("   → Statistically significant at α=0.05")
else:
    print("   → Not statistically significant at α=0.05")
```

- **From r value:** Countries with higher R&D spending (% of GDP) tend to have more patents, but the relationship isn't tight.
- **From p value:** This pattern is very unlikely to be due to random chance in the data sample.
- **$r^2 = 0.124$ says:** Only ~12% of the variation in patent counts is explained by R&D spending alone.

```python
from scipy.stats import spearmanr

spearman_r, spearman_p = spearmanr(df_B_agg['rd_gdp'], df_B_agg['patent_count'])

print(f" Spearman rho = {spearman_r:.3f}")
print(f" Spearman p-value = {spearman_p:.3f}")
print(f" Difference in Spearman rho and Pearson r value = {abs(spearman_r - r):.3f}")
```

- Strong evidence of a **MONOTONIC** relationship : So as R&D spending increases, patent output would also inscrease.
- But the relationship is **NON-LINEAR** :Patent counts are highly skewed as most countries have few patents; a few have thousands. 

```python
import numpy as np

df_B_agg['log_patents'] = np.log1p(df_B_agg['patent_count'])

# to compare the plots side-by-side. 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.regplot(data=df_B_agg, x='rd_gdp', y='patent_count', 
            scatter_kws={'alpha':0.6, 's':40}, line_kws={'color':'red'}, ax=axes[0])
axes[0].set_xlabel('R&D Spending (% of GDP)')
axes[0].set_ylabel('Total Patent Count')
axes[0].set_title('Raw Scale')
axes[0].grid(alpha=0.3)

sns.regplot(data=df_B_agg, x='rd_gdp', y='log_patents', 
            scatter_kws={'alpha':0.6, 's':40}, line_kws={'color':'green'}, ax=axes[1])
axes[1].set_xlabel('R&D Spending (% of GDP)')
axes[1].set_ylabel('Log(Patent Count + 1)')
axes[1].set_title('Log-Transformed Scale')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

r_log, p_log = pearsonr(df_B_agg['rd_gdp'], df_B_agg['log_patents'])

print(f" Pearson r = {r:.3f}")
print(f" Pearson r (log patents) = {r_log:.3f}")
print(f" Spearman rho = {spearman_r:.3f}")
print(f" p-value (log-transformed) = {p_log:.3f}")
```

- **r values:** Strong linear correlation after log-transform
- **p value(log and regular):** Statistically significant
- Countries with higher R&D spending (% of GDP) tend to have exponentially more patents.
- A small increase in R&D % is associated with a multiplicative increase in patent output.

### Is there a lag effect? 


- For this we will test this by adding lag to each step and test out the resulting effects in comparison to the data.

```python
df_agg = df_B_agg.groupby(['country_code', 'year'], observed=True).agg({
    'patent_count': 'sum',
    'inventor_count': 'sum',
    'rd_gdp': 'first',
    'gdp_growth': 'first', 
    'population': 'first',
    'tertiary_enrollment': 'first',
    'log_patents': 'sum'  
}).reset_index()

df_agg = df_agg.sort_values(['country_code', 'year']).reset_index(drop=True)

print(df_agg[df_agg['country_code'] == 'ARE'][['year', 'patent_count', 'log_patents']].head(10))
```

##### There are gap years as can be seen, so we need to handle it via shift(). That way, lags will be pulled from years right before. 

```python
df_agg['log_patents_lag1'] = df_agg.groupby('country_code', observed=True)['log_patents'].shift(1)

print(df_agg[df_agg['country_code'] == 'ARE'][['year', 'log_patents', 'log_patents_lag1']].head(10))
```

```python
df_clean = df_agg.dropna(subset=['log_patents_lag1'])

pearson_r, pearson_p = pearsonr(df_clean['log_patents_lag1'], df_clean['gdp_growth'])
spearman_r, spearman_p = spearmanr(df_clean['log_patents_lag1'], df_clean['gdp_growth'])

print(f"Pearson r: {pearson_r:.4f}, p-value: {pearson_p:.4f}")
print(f"Spearman rho: {spearman_r:.4f}, p-value: {spearman_p:.4f}")
print(f"Sample size: {len(df_clean)}")
```

##### 1 year lag cannot comprehend the resulting effects fully as it seems.
- **r** values are still negative.
So we will try larger lag time .

```python
for lag in [2, 3, 4, 5]:
    df_agg[f'log_patents_lag{lag}'] = df_agg.groupby('country_code', observed=True)['log_patents'].shift(lag)

results = []
for lag in [1, 2, 3, 4, 5]:
    col = f'log_patents_lag{lag}'
    df_temp = df_agg.dropna(subset=[col, 'gdp_growth'])
    r, p = spearmanr(df_temp[col], df_temp['gdp_growth']) 
    results.append({'lag': lag, 'spearman_r': r, 'p_value': p, 'n': len(df_temp)})

pd.DataFrame(results)
```

- The negative correlation persists across all lags, though it weakens slightly.
- We speculate two likely scenarios:
  + Maybe GDP growth drives patenting activity!
  + The relationship might be positive for high-income countries but negative for low-income ones and averaging masks this.

#### For the 1st scenario: 

```python
df_agg['gdp_growth_lag1'] = df_agg.groupby('country_code', observed=True)['gdp_growth'].shift(1)

df_clean = df_agg.dropna(subset=['gdp_growth_lag1', 'log_patents'])
r, p = spearmanr(df_clean['gdp_growth_lag1'], df_clean['log_patents'])

print(f"Spearman rho: {r:.4f}, p-value: {p:.4f}, n: {len(df_clean)}")
```

##### This contradicts the 1st scenario. Because this r value is also negative. So GDP is not driving patenting activity generally.


#### For 2nd scenario we separate into high income come spending and low income come spending group.

```python
df_agg['rd_group'] = df_agg['rd_gdp'].apply(lambda x: 'High_RnD' if x >= 1.0 else 'Low_RnD')

for group in ['High_RnD', 'Low_RnD']:
    subset = df_agg[df_agg['rd_group'] == group].dropna(subset=['log_patents_lag1', 'gdp_growth'])
    r, p = spearmanr(subset['log_patents_lag1'], subset['gdp_growth'])
    print(f"{group}: rho = {r:.4f}, p = {p:.4f}, n = {len(subset)}")
```

#### So the negative correlation is not an averaging effect as for both cases we found r to be negative. 
- As the negativity was shrinking with larger lag, we'll now test the hypothesis of large lag effects!

```python
for lag in range(7, 13):
    df_agg[f'log_patents_lag{lag}'] = df_agg.groupby('country_code', observed=True)['log_patents'].shift(lag)

results_long = []
for lag in range(7, 13):
    col = f'log_patents_lag{lag}'
    df_temp = df_agg.dropna(subset=[col, 'gdp_growth'])
    
    if len(df_temp) < 100:  
        continue
        
    r, p = spearmanr(df_temp[col], df_temp['gdp_growth'])
    results_long.append({
        'lag': lag, 
        'spearman_r': r, 
        'p_value': p, 
        'n': len(df_temp)
    })

pd.DataFrame(results_long)
```

#### Key Findings from this:
- The negetivity shrunk than at 12 year mark, it jumped up a bit again!
- For now we may conclude:
    + **Short-term:** Patenting is costly (R&D spending, resource allocation) that drags on measured GDP growth
    + **Long-term:** The drag fades, but the positive payoff might be too diffuse or conditional to show up in a simple correlation

#### As we couldn't conclude with direct indications from lag effects in patent output to GDP growth, we will shift our focus now towards the R&D spending to patent output.
- It had much stronger and positive correlations from the start.
- It might indicate towards a possible decision that might help us pin down the pattern driving the overall picture.

```python
results2 = []

for lag in [0, 1, 2, 3, 4, 5]:
    if lag == 0:
        df_temp = df_agg.dropna(subset=['rd_gdp', 'log_patents'])
        r, p = spearmanr(df_temp['rd_gdp'], df_temp['log_patents'])
    else:
        patent_col = f'log_patents_lag{lag}'
        df_temp = df_agg.dropna(subset=['rd_gdp', patent_col])
        r, p = spearmanr(df_temp['rd_gdp'], df_temp[patent_col])
    
    results2.append({
        'lag_years': lag,
        'spearman_r': r,
        'p_value': p,
        'n': len(df_temp)
    })

pd.DataFrame(results2)
```

- **Primary finding:** R&D spending (% of GDP) shows a strong, positive monotonic association with patent output (Spearman ρ = 0.81, p < 0.001). This relationship remains stable across 0-5 year lags, suggesting patent production responds to R&D investment without a detectable multi-year delay at annual resolution.
- **Secondary finding:** The association between patent output and GDP growth is weakly negative across all tested lags (1-12 years), with the magnitude decaying over time (ρ = -0.25 at 1-year → -0.09 at 12-year). This suggests that while patenting may entail short-term economic costs, any long-term growth payoff is either diffuse, conditional on other factors, or not captured by simple bivariate correlation.
- We have not tested for all other indicators and factors(population, education, tertiary enrollment, separation by types of patents, socioeconomic variables, macroeconomic indicators etc.), as it may go to complexity levels way beyond the goals of this study, but could surely be done so as per necessity. 

### Which countries convert R&D budget -> patents most efficiently?


- For the mathematical formulation to calculate the efficiency, we can follow this derivation that I thought was consistent with our data format. 
- Efficiency, $$E = \frac{\frac{\text{patent\_count}}{\text{population}}}{\frac{\text{rd\_gdp} \times \text{inventor\_count}}{\text{population}}} \times 10^{6}$$

- $\implies$ $$E = \frac{\text{patent\_count}}{\text{rd\_gdp} \times \text{inventor\_count}} \times 10^{6}$$

```python
print(df_B_agg.columns.tolist())
print(df_B_agg.shape)
```

- Too low patent_count, inventor_count, rd_gdp might give anomalous results. So, filtering out by setting minimal conditions is the approach we go with.

```python
MIN_PATENTS = 500           
MIN_INVENTORS = 2000          
MIN_RD_GDP = 1            
MIN_POPULATION = 5_000_000  

df_filtered = df_B_agg[
    (df_B_agg['patent_count'] >= MIN_PATENTS) &
    (df_B_agg['inventor_count'] >= MIN_INVENTORS) &
    (df_B_agg['rd_gdp'] >= MIN_RD_GDP) &
    (df_B_agg['population'] >= MIN_POPULATION)
].copy()

print(f"Rows before filter: {len(df_B_agg)}")
print(f"Rows after filter: {len(df_filtered)}")
print(f"Countries retained: {df_filtered['country_code'].nunique()}")
print(f"Years covered: {df_filtered['year'].min()} to {df_filtered['year'].max()}")
```

```python
df_filtered['Efficiency_1'] = (
    df_filtered['patent_count'] / 
    (df_filtered['rd_gdp'] * df_filtered['inventor_count'] )
) * 1e6


ranked = df_filtered.groupby('country_code', observed=True).agg({
    'Efficiency_1': 'mean',     
    'patent_count': 'sum',    
    'rd_gdp': 'mean',         
    'inventor_count': 'mean'   
}).round(2)

print(" Top 10 Countries by R&D→Patent Efficiency :")
print(ranked.sort_values('Efficiency_1', ascending=False).head(10))
```

- We see largely varying results just by changing the filtering conditions!
- We'll finally have to take reasonable conditions to filter. 

### Which countries convert patent output -> GDP growth most efficiently? 

##### Similarly to the other efficiency, we can follow this formula for calculating efficiency of Countries converting patent output to GDP growth most efficiently.

$$E = \frac{\text{gdp\_growth}}{\frac{\text{patent\_count}}{\text{population}}} \times 10^{6}$$

```python
print(df_B_agg.columns.tolist())
print(df_B_agg.shape)
```

```python
MIN_PATENTS = 500           
MIN_INVENTORS = 2000          
MIN_RD_GDP = 1            
MIN_POPULATION = 5_000_000  

df_filtered = df_B_agg[
    (df_B_agg['patent_count'] >= MIN_PATENTS) &
    (df_B_agg['inventor_count'] >= MIN_INVENTORS) &
    (df_B_agg['rd_gdp'] >= MIN_RD_GDP) &
    (df_B_agg['population'] >= MIN_POPULATION)
].copy()

print(f"Rows before filter: {len(df_B_agg)}")
print(f"Rows after filter: {len(df_filtered)}")
print(f"Countries retained: {df_filtered['country_code'].nunique()}")
print(f"Years covered: {df_filtered['year'].min()} to {df_filtered['year'].max()}")
```

```python
df_filtered['Efficiency_2'] = (
    df_filtered['gdp_growth'] / 
    (df_filtered['patent_count'] / df_filtered['population'] )
) * 1e6


ranked = df_filtered.groupby('country_code', observed=True).agg({
    'Efficiency_2': 'mean',     
    'patent_count': 'sum',    
    'rd_gdp': 'mean',         
    'inventor_count': 'mean'   
}).round(2)

print(" Top 10 Countries by Patent -> GDP Growth Efficiency :")
print(ranked.sort_values('Efficiency_2', ascending=False).head(10))
```

```python
def assign_group(code):
    if code in HIGH_INCOME:
        return 'High Income'
    elif code in EMERGING_ECONOMIES:
        return 'Emerging'
    else:
        return 'Other'

df_filtered['group'] = df_filtered['country_code'].apply(assign_group)
```

```python
yearly = (
    df_filtered[df_filtered['group'] != 'Other']
    .groupby(['year', 'group'],observed=True)[['Efficiency_1', 'Efficiency_2']]
    .mean()
    .reset_index()
)

hi = yearly[yearly['group'] == 'High Income']
em = yearly[yearly['group'] == 'Emerging']
```

### Does R&D patent volume plateau after certain R&D thresholds? 

```python
df_filtered['rd_bin'] = pd.qcut(df_filtered['rd_gdp'].dropna(), q=10, duplicates='drop')

threshold_df = (
    df_filtered[df_filtered['group'] != 'Other']
    .groupby(['rd_bin', 'group'], observed=False)['patent_count']
    .mean()
    .reset_index()
)

threshold_df['rd_bin_mid'] = threshold_df['rd_bin'].apply(lambda x: x.mid)
threshold_df = threshold_df.sort_values('rd_bin_mid')
```

```python
fig, ax = plt.subplots(figsize=(9, 4))

for group, color in [('High Income', 'steelblue'), ('Emerging', 'seagreen')]:
    subset = threshold_df[threshold_df['group'] == group]
    ax.plot(subset['rd_bin_mid'], subset['patent_count'],
            marker='o', label=group, color=color)

ax.set_xlabel('R&D spend (% of GDP) — binned midpoint')
ax.set_ylabel('Mean patent count')
ax.set_title('Patent output vs R&D spend — plateau detection')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

def log_model(x, a, b):
    return a * np.log(x) + b

hi_clean = df_filtered[df_filtered['group'] == 'High Income'].dropna(subset=['rd_gdp', 'patent_count'])
hi_clean = hi_clean[hi_clean['rd_gdp'] > 0]

x = hi_clean['rd_gdp'].values
y = hi_clean['patent_count'].values

popt, _ = curve_fit(log_model, x, y, maxfev=5000)

x_range = np.linspace(x.min(), x.max(), 300)
y_hat = log_model(x_range, *popt)

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(x, y, alpha=0.3, s=15, color='steelblue', label='High Income (observed)')
ax.plot(x_range, y_hat, color='crimson', linewidth=2, label=f'Log fit: {popt[0]:.2f}·ln(x) + {popt[1]:.2f}')
ax.set_xlabel('R&D spend (% of GDP)')
ax.set_ylabel('Patent count')
ax.set_title('Log curve fit — High Income: R&D vs Patent output')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
