```python
import pandas as pd

chunk_size = 500_000
null_counts = None
total_rows = 0

for i, chunk in enumerate(pd.read_csv("../raw_data/g_patent.tsv", sep="\t", chunksize=chunk_size, low_memory=False)):
    total_rows += len(chunk)
    
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    
    print(f"Chunk {i+1} processed | Rows so far: {total_rows:,}")

print(f"\nTotal rows: {total_rows:,}")
print("\nNull counts per column:")
print(null_counts)
```

#### There is no country code or country name found for the inventors or the patents. So I searched up the **USPTO** website and found a file that is clearly referenced for these and other datasets to get *country name* out from with documentation called `g_location_disambiguated.tsv`. So I went ahead and downloaded that manually from the website. 

```python
locations = pd.read_csv('../raw_data/g_location_disambiguated.tsv', sep='\t', quotechar='"',
                        usecols=['location_id', 'disambig_country'])
location_lookup = locations.set_index('location_id')['disambig_country'].to_dict()
total_rows = 0
all_countries = set()
for i, chunk in enumerate(pd.read_csv("../raw_data/g_inventor_disambiguated.tsv", sep="\t", chunksize=500_000, low_memory=False)):
    total_rows += len(chunk)
    chunk['country'] = chunk['location_id'].map(location_lookup)
    all_countries.update(chunk['country'].dropna().unique())
print(f"\nTotal rows: {total_rows:,}")
print(f"Total unique countries: {len(all_countries)}")
sorted_countries = sorted(all_countries)
print(sorted_countries)
```
